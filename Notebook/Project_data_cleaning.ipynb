### Handling Negative Weekly Sales
There are 1285 total rows showing negative sales. out of which there are 45 unique stores, 50 different dept, 143 unique dates, 678 sales value, and 2 holidays meaning returns have been done during holiday weeks and during non holiday weeks both.The negative sales could be refunds or returns. since we are forecasting demands I would remove the negative sales as they are nott the actual demands and also these are just 0.3% of all sales
